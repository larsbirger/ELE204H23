# Problem 1

Given the following $Y(s)$ that represents the unit step response of a dynamical system:

$$Y(s)=\frac{s+12}{s(s+3)}=\frac{c_1}{s}+\frac{c_2}{s+3}$$

## a) 

Use partial fraction expansion (*delbrøksoppløsning*) and find the coefficients $c_1$ and $c_2$;

```python
import sympy as sym
s, c_1, c_2 = sym.symbols('s,c_1,c_2')

Y = sym.Eq((s+12)/(s*(s+3)),c_1/s + c_2/(s+3))
display(Y)

sym.solve(Y,(c_1,c_2))
```

## b)

Use the inverse Laplace transform from the Laplace-table and determine $y(t)$;

utkast av tabellen:

| $s$-plane       | $t$-plane  |
| --------------- | ---------- |
| $\frac{k}{s}$   | $k$        |
| $\frac{k}{a+s}$ | $ke^{-at}$ |

since the equation is:
$$\frac{4}{s}-\frac{3}{s+3}$$

and since Laplace is a linear operator we can split the terms

$\frac{4}{s}$ and $-\frac{3}{s+3}$

$$\frac{4}{s} = \mathscr{L}\{4\}$$

$$-\frac{3}{s+3} = \mathscr{L}\{-3e^{-3t}\}$$

and here is a script for verification:

```python
import sympy as sym
s,t = sym.symbols('s,t')
Y = sym.apart((s+12)/(s*(s+3)))
display(Y)


y = sym.simplify(
        sym.inverse_laplace_transform(Y,s,t)
        /
        sym.Heaviside(t))

display(y)
```

$$W(s)=\frac{H(s)}{G(s)}$$

# Problem 3

Consider the electrical circuit shown below with the output $i_L(t)$ and the input $i_G(t)$. The differential equations describing the reactive elements are $u_L(t) = L\frac{di_L(t)}{dt} = L{i'}_L(t)$, $i_C(t) = C\frac{du_C(t)}{dt} = C{u'}_C(t)$. The initial values are $i_L(0) = 0\;[A]$

## a)

Find the differential equation of the circuit;

Lets start with the KCL of the node where they explicitly placed a voltage reference point:
$$i_G=i_R+i_C+i_L$$

Now, there are two variables we don't care about here, namely $i_R$ and $i_C$, and two variables match with the specified input ($i_G$) and output ($i_L$).
lets see if we can find some other ways to describe these two unwanted variables.
$$i_R=\frac{u_R}{R}$$
$$i_C=\frac{1}{C}\frac{du_C}{dt}$$

and for general housekeeping:
$$i_L=i_{R_L}$$


This makes the KCL look like this:
$$i_G=\frac{u_R}{R}+\frac{1}{C}\frac{du_C}{dt}+i_L$$

We can now point out how the node has the same voltage, and since all branches share this node and another common node we can say:
$$u=u_R=u_C=u_L+u_{R_L}$$

This concludes:
$$u=u_L+u_{R_L}$$
$$u=L\frac{di_L}{dt}+R_Li_L$$

and since the KCL equation needs $\frac{du}{dt}$ we can derive $u$
$$u=L\frac{di_L}{dt}+R_Li_L$$
$$\frac{du}{dt}=L\frac{d^2i_L}{(dt)^2}+R_L\frac{di_L}{dt}$$

Making the equation
$$i_G=\frac{u}{R}+\frac{1}{C}\frac{du}{dt}+i_L$$

Look like this:
$$i_G=\frac{L\frac{di_L}{dt}+R_Li_L}{R}+\frac{1}{C}(L\frac{d^2i_L}{(dt)^2}+R_L\frac{di_L}{dt})+i_L$$

When factoring it:
$$i_G=\frac{L\frac{di_L}{dt}}{R}+\frac{R_Li_L}{R}+\frac{1}{C}L\frac{d^2i_L}{(dt)^2}+\frac{1}{C}R_L\frac{di_L}{dt}+i_L$$

$$i_G=\frac{L}{R}\frac{di_L}{dt}+\frac{R_L}{R}i_L+\frac{L}{C}\frac{d^2i_L}{(dt)^2}+\frac{R_L}{C}\frac{di_L}{dt}+i_L$$

$$i_G=\frac{L}{R}\frac{di_L}{dt}+\frac{R_L}{R}i_L+\frac{L}{C}\frac{d^2i_L}{(dt)^2}+\frac{R_L}{C}\frac{di_L}{dt}+i_L$$

Concludes this:
$$i_G=\frac{L}{C}\frac{d^2i_L}{(dt)^2}+(\frac{L}{R}+\frac{R_L}{C})\frac{di_L}{dt}+(\frac{R_L}{R}+1)i_L$$

which seems to be the differential equation that explains the system.

## b)

Find the transfer function of the circuit;

The transfer function ($W(s)$) is on the shape:
$$W(s)=\frac{O_{ut}(s)}{I_{nn}(s)}$$

According to the problem description $O_{ut} = I_L$ and $I_{nn} = I_G$ making the transfer function be on the shape:
$$W(s)=\frac{I_L}{I_G}$$

Lets make our differential equation enter the $S$-plane:
$$\mathscr{L}\big\{i_G\big\}=\mathscr{L}\bigg\{\frac{L}{C}\frac{d^2i_L}{(dt)^2}+(\frac{L}{R}+\frac{R_L}{C})\frac{di_L}{dt}+(\frac{R_L}{R}+1)i_L\bigg\}$$

the initial values equals 0 so i will deliberately ignore them now.

$$I_G=\frac{L}{C}s^2I_L+(\frac{L}{R}+\frac{R_L}{C})sI_L+(\frac{R_L}{R}+1)I_L$$

Then we can algebraically isolate for $\frac{I_L}{I_G}$:
$$I_G=\frac{L}{C}s^2I_L+(\frac{L}{R}+\frac{R_L}{C})sI_L+(\frac{R_L}{R}+1)I_L$$

$$\frac{I_G}{I_L}=\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)$$

$$\frac{I_L}{I_G}=\frac{1}{\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)}$$

Finally, we arrive at the transfer function:
$$W(s)=\frac{1}{\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)}$$

## c)

What is the output when $t\rightarrow\infty$ and the input is a step with amplitude $I_G\;[A]$?

This is a circuit with passive components, resulting in a positive dampening ratio, so it is stable, making the final value theorem apply.

$$\lim_{t\rightarrow\infty}{f(t)} = \lim_{s\rightarrow 0}{sF(s)}$$

Input is now a unit step of amplitude $i_G$ resulting in $\frac{I_G}{s}$


$$\frac{I_L}{\frac{I_G}{s}}=\frac{1}{\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)}$$

$$I_L=\frac{\frac{I_G}{s}}{\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)}$$

$$\lim_{s \rightarrow 0} s \frac{\frac{I_G}{s}}{\frac{L}{C}s^2+(\frac{L}{R}+\frac{R_L}{C})s+(\frac{R_L}{R}+1)}$$

```python
import sympy as sp
s,IG,L,C,R,RL = sp.symbols("s,I_G,L,C,R,R_L")


IL = (IG/s)/((L/C)*s**2+(L/R+RL/C)*s+(RL/R+1))
display(IL)
display((s*IL).subs(s,0))
display(sp.simplify((s*IL).subs(s,0)))
```

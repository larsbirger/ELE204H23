# Problem 1

$$Y(s)=\frac{s+12}{s(s+3)}=\frac{c_1}{s}+\frac{c_2}{s+3}$$

## a) 

```python
import sympy as sym
s, c_1, c_2 = sym.symbols('s,c_1,c_2')

Y = sym.Eq((s+12)/(s*(s+3)),c_1/s + c_2/(s+3))
display(Y)

sym.solve(Y,(c_1,c_2))
```

## b)

```python
import sympy as sym
s,t = sym.symbols('s,t')
Y = sym.apart((s+12)/(s*(s+3)))
display(Y)

def ilaplace(W):
    import sympy as sym
    s,t = sym.symbols('s,t')
    return sym.simplify(sym.inverse_laplace_transform(W,s,t)/sym.Heaviside(t))

y = ilaplace(Y)
display(y)
```

$\theta(t)$ is the Heavyside function, it is there to signify that this result only makes sense from $t\geq0$
